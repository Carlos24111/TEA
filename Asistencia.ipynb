```python
# Asistencia en clase en porcentaje 
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
asist = pd.read_csv("asistencia.csv")
asist  
```

```python
#Asist [["temp", "germinated"]] 
asist[["Código", "Asistió", "No asistió"]] 
# agrupado por periodo de tiempo y sumando lo germinado en ese periodo
df_a = asist.groupby(["Asistió", "No asistió"], as_index=False, sort=False)
df_a   
```

```python
asist.drop(["Unnamed: 3", "Unnamed: 4","Unnamed: 5"], axis= 1) 
  
```

```python
Presencia = asist["Asistió"].sum() 
Presencia
```

```python
Asistió = Presencia
Asistió 
```

```python
Ausencia = asist["No asistió"].sum() 
Ausencia  
```

```python
No_asistió = Ausencia 
No_asistió 
```

```python
Total = Asistió + No_asistió 
Total  
```

```python
p = [["Asistió",Asistió], ["No_asistió",No_asistió]] 
df = pd.DataFrame(p) 
df    
```

```python
df = pd.DataFrame(p, columns=["Asistencia", "Total"]) 
df      
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt    
```

```python
colors = ['blue', 'red', 'green']  
# Plotting the pie chart for above dataframe
df.groupby(['Asistencia']).sum().plot(
    kind='pie', ylabel='Total', subplots = True,
  autopct='%1.0f%%', colors=colors)    
plt.show()   
```

```python
colors = ['blue', 'red', 'green']  
# Plotting the pie chart for above dataframe
df.groupby(['Asistencia']).sum().plot(
    kind='pie', subplots = True,
  autopct='%1.1f%%', colors=colors, startangle = 140)    
plt.show()      
```
