```python
import matplotlib.pyplot as plt
import pandas as pd
Asistencia = percentage["Asistencia"]
Total = percentage["Total"]
colors = ["#1f77b4", "#ff7f0e"]
explode = (0.1, 0)  
plt.pie(Total, labels=Asistencia, explode=explode, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Grafico de porcentajes de asistencia")
plt.show()
```
